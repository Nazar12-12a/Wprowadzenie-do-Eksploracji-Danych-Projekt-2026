```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# ==========================================
# KONFIGURACJA (TUTAJ ZMIENIASZ ROK I FOLDER)
# ==========================================
rok_analizy = 2013
nazwa_folderu = 'wykresy_turystyka'  # Tu trafi wszystkie wykresy

# Tworzenie folderu (jeli nie istnieje)
os.makedirs(nazwa_folderu, exist_ok=True)

# 1. Wczytanie danych
df = pd.read_csv('master_dataset_project.csv')

# 2. Filtrowanie (TOP 10 dla wybranego roku)
df_rok = df[df['Rok'] == rok_analizy].sort_values(by='Udzielone_Noclegi', ascending=False).head(10)

# 3. Rysowanie wykresu
plt.figure(figsize=(14, 8))
sns.set_style("whitegrid")

# Wykres supkowy
ax = sns.barplot(data=df_rok, x='Udzielone_Noclegi', y='Nazwa', palette='Blues_r')

# Dodanie liczb na supkach (etykiety danych)
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', padding=5, fontsize=11, fontweight='bold')

# Tytuy i opisy (Po polsku)
plt.title(f'TOP 10 Powiatw z najwiksz liczb noclegw w {rok_analizy} roku', fontsize=16, fontweight='bold')
plt.xlabel('Liczba udzielonych noclegw (szt.)', fontsize=12)
plt.ylabel('Powiat', fontsize=12)

# rdo danych (na dole po prawej)
plt.figtext(0.95, 0.02, "rdo danych: BDL GUS", ha="right", fontsize=10, style='italic', color='gray')

# 4. Zapis do folderu
nazwa_pliku = f'top10_powiatow_{rok_analizy}.png'
sciezka_zapisu = os.path.join(nazwa_folderu, nazwa_pliku)

plt.tight_layout()
plt.savefig(sciezka_zapisu, dpi=300) # Wysoka jako
plt.show()

print(f" Wykres zapisano w folderze '{nazwa_folderu}': {nazwa_pliku}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# ==========================================
# KONFIGURACJA ()
# ==========================================
nazwa_folderu = 'wykresy_turystyka'
os.makedirs(nazwa_folderu, exist_ok=True) # Tworzy folder /  

# 1. Wczytanie danych
df = pd.read_csv('master_dataset_project.csv')

# Lista wszystkich dostpnych lat w pliku
lata = sorted(df['Rok'].unique())

print(f"Rozpoczynam generowanie wykresw dla lat: {lata}...")

# ==========================================
# PTLA GENERUJCA WYKRESY ()
# ==========================================
for rok in lata:
    # 2. Filtrowanie TOP 10 dla danego roku
    df_rok = df[df['Rok'] == rok].sort_values(by='Udzielone_Noclegi', ascending=False).head(10).copy()
    
    # Zmiana nazw na Wielkie Litery (np. "powiat tatrzaski" -> "Powiat Tatrzaski")
    df_rok['Nazwa'] = df_rok['Nazwa'].str.title()
    
    # 3. Rysowanie
    plt.figure(figsize=(14, 8))
    sns.set_style("whitegrid")
    
    # Wykres
    ax = sns.barplot(data=df_rok, x='Udzielone_Noclegi', y='Nazwa', palette='Blues_r')
    
    # Etykiety z liczbami (czcionka pogrubiona)
    for container in ax.containers:
        ax.bar_label(container, fmt='%.0f', padding=5, fontsize=11, fontweight='bold')

    # Tytu i osie
    plt.title(f'TOP 10 Powiatw: Liczba udzielonych noclegw w {rok} roku', fontsize=16, fontweight='bold')
    plt.xlabel('Liczba noclegw [szt.]', fontsize=12)
    plt.ylabel('Powiat', fontsize=12)
    
    # rdo danych (prawy dolny rg)
    plt.figtext(0.95, 0.02, "rdo danych: BDL GUS", ha="right", fontsize=10, style='italic', color='gray')

    # 4. Zapis do pliku
    nazwa_pliku = f'top10_noclegi_{rok}.png'
    sciezka = os.path.join(nazwa_folderu, nazwa_pliku)
    
    plt.tight_layout()
    plt.savefig(sciezka, dpi=300) # Wysoka jako
    plt.close() # Zamknij wykres, eby nie zapycha pamici
    
    print(f" Zapisano: {nazwa_pliku}")

print(f"\n Gotowe! Wszystkie {len(lata)} wykresw znajdziesz w folderze '{nazwa_folderu}'.")
```
