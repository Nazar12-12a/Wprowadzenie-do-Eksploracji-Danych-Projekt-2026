```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
import os

# --- USTAWIENIA STYLU ---
sns.set_theme(style="whitegrid", context="talk")
plt.rcParams['font.family'] = 'Arial'

# 1. WCZYTANIE DANYCH
file_name = "master_dataset_project.csv"

if os.path.exists(file_name):
    df = pd.read_csv(file_name)
    print(f"Wczytano plik: {file_name}")
elif os.path.exists(r"C:\Nazar Haievskyj\Dane w\master_dataset_project.csv"):
    df = pd.read_csv(r"C:\Nazar Haievskyj\Dane w\master_dataset_project.csv")
else:
    raise FileNotFoundError("Brak pliku master_dataset_project.csv")

# Czyszczenie
df.columns = df.columns.str.strip()
df['Rok'] = pd.to_numeric(df['Rok'], errors='coerce')
df = df.dropna(subset=['Rok'])
df['Rok'] = df['Rok'].astype(int)

# =========================================================
# WYKRES TRENDU (2013 - 2024)
# =========================================================

trend_data = df.groupby('Rok')['Udzielone_Noclegi'].sum().reset_index()
trend_data['Noclegi_mln'] = trend_data['Udzielone_Noclegi'] / 1_000_000

plt.figure(figsize=(16, 9))

# Linia g≈Ç√≥wna
plt.plot(trend_data['Rok'], trend_data['Noclegi_mln'], 
         marker='o', markersize=9, color='#003366', linewidth=3, label='Ruch turystyczny')

# COVID-19
covid_period = trend_data[(trend_data['Rok'] >= 2019) & (trend_data['Rok'] <= 2021)]
plt.plot(covid_period['Rok'], covid_period['Noclegi_mln'], 
         color='#cc0000', linewidth=4, label='Pandemia COVID-19')

# Rok 2024
last_point = trend_data[trend_data['Rok'] == 2024]
if not last_point.empty:
    plt.scatter(last_point['Rok'], last_point['Noclegi_mln'], color='#28a745', s=150, zorder=5, label='Rok 2024')

# Osie i Tytu≈Çy
plt.gca().yaxis.set_major_formatter(ticker.FormatStrFormatter('%.0f'))
plt.title("Dynamika udzielonych nocleg√≥w w Polsce (2013‚Äì2024)", fontsize=22, fontweight='bold', pad=20)
plt.xlabel("Rok", fontsize=14, fontweight='bold')
plt.ylabel("Liczba udzielonych nocleg√≥w [mln szt.]", fontsize=14, fontweight='bold')
plt.xticks(trend_data['Rok'].unique()) 
plt.legend(fontsize=12, loc='upper left')
plt.grid(True, alpha=0.3)
plt.figtext(0.95, 0.02, "≈πr√≥d≈Ço: Opracowanie w≈Çasne na podstawie danych BDL GUS", ha="right", fontsize=10, style='italic')

plt.tight_layout()

# --- ZAPISYWANIE PLIKU ---
# To jest kluczowa linijka, kt√≥ra zapisuje obrazek na dysku
nazwa_pliku = "Wykres_Trend_2013_2024.png"
plt.savefig(nazwa_pliku, dpi=300, bbox_inches='tight')

print(f"Sukces! Wykres zosta≈Ç zapisany jako plik: {nazwa_pliku}")
plt.show()
```
