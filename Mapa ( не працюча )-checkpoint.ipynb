```python
import pandas as pd
import folium
import requests
import json
import branca.colormap as cm

# ==========================================
# 1. WCZYTANIE DANYCH (Ð—ÐÐ’ÐÐÐ¢ÐÐ–Ð•ÐÐÐ¯)
# ==========================================
print("ðŸ”„ WczytujÄ™ TwÃ³j plik Master Dataset...")

# UÅ¼ywamy Twojego gotowego pliku (nie zmieniamy go, tylko czytamy)
df = pd.read_csv('master_dataset_project.csv')

# Wybieramy najnowszy rok
ostatni_rok = df['Rok'].max()
print(f"ðŸ“… Rok danych do mapy: {ostatni_rok}")

# Tworzymy kopiÄ™ roboczÄ… tylko dla mapy
df_mapa = df[df['Rok'] == ostatni_rok].copy()

# ==========================================
# 2. TRANSFORMACJA KODÃ“W (STRATEGIA LICZBOWA)
# ==========================================
print("ðŸ”§ Przetwarzam kody na liczby (Integer Matching)...")

# Funkcja zamieniajÄ…ca kod BDL (np. 201000) na prostÄ… liczbÄ™ ID powiatu (np. 201)
def przygotuj_id_powiatu(kod_z_pliku):
    # 1. ZamieÅ„ na tekst i dodaj wiodÄ…ce zera (Å¼eby mieÄ‡ 7 znakÃ³w: "0201000")
    s = str(kod_z_pliku).zfill(7)
    # 2. WeÅº pierwsze 4 znaki ("0201") - to jest wÅ‚aÅ›ciwy kod powiatu
    kod_4_znaki = s[:4]
    # 3. ZamieÅ„ na liczbÄ™ caÅ‚kowitÄ… (usuwa zera z przodu: "0201" -> 201)
    return int(kod_4_znaki)

# Zastosowanie funkcji
df_mapa['Match_ID'] = df_mapa['Kod'].apply(przygotuj_id_powiatu)

# Poprawa nazw (Wielkie Litery)
df_mapa['Nazwa'] = df_mapa['Nazwa'].str.title()

# Tworzymy sÅ‚owniki: Klucz to LICZBA (np. 201), WartoÅ›Ä‡ to Noclegi/Nazwa
dict_noclegi = df_mapa.set_index('Match_ID')['Udzielone_Noclegi'].to_dict()
dict_nazwy = df_mapa.set_index('Match_ID')['Nazwa'].to_dict()

# ==========================================
# 3. POBRANIE I DOPASOWANIE MAPY
# ==========================================
print("ðŸŒ Pobieram mapÄ™ GeoJSON...")
url_geojson = "https://raw.githubusercontent.com/ppatrzyk/polska-geojson/master/powiaty/powiaty-min.geojson"
geo_data = requests.get(url_geojson).json()

matched = 0
total = 0

for feature in geo_data['features']:
    total += 1
    # Pobieramy ID z mapy i teÅ¼ zamieniamy na liczbÄ™ (int)
    # Mapa moÅ¼e mieÄ‡ "0201" lub "201" -> w obu przypadkach int() da 201
    map_id_raw = feature['properties']['id']
    try:
        map_id_int = int(map_id_raw)
    except:
        map_id_int = -1 # BÅ‚Ä…d konwersji (rzadki przypadek)

    # Szukamy po liczbie! (201 == 201)
    if map_id_int in dict_noclegi:
        wartosc = dict_noclegi[map_id_int]
        nazwa = dict_nazwy[map_id_int]
        matched += 1
    else:
        wartosc = 0
        nazwa = feature['properties']['nazwa'] # Zapasowa nazwa z mapy

    # Zapisujemy dane do mapy
    feature['properties']['Noclegi'] = int(wartosc)
    feature['properties']['Nazwa_Wyswietlana'] = nazwa

print(f"âœ… SUKCES: Dopasowano {matched} z {total} powiatÃ³w.")

# ==========================================
# 4. RYSOWANIE MAPY
# ==========================================
print("ðŸŽ¨ GenerujÄ™ plik HTML...")

m = folium.Map(location=[52.0, 19.1], zoom_start=6, tiles="cartodbpositron")

# Skala kolorystyczna (95 centyl dla kontrastu)
max_val = df_mapa['Udzielone_Noclegi'].quantile(0.95)

colormap = cm.LinearColormap(
    colors=['#ffffff', '#c6dbef', '#6baed6', '#3182bd', '#08519c'], 
    index=[0, max_val*0.1, max_val*0.3, max_val*0.6, max_val],
    vmin=0, vmax=max_val,
    caption=f'Liczba udzielonych noclegÃ³w ({ostatni_rok})'
)

folium.GeoJson(
    geo_data,
    style_function=lambda feature: {
        'fillColor': colormap(feature['properties']['Noclegi']) if feature['properties']['Noclegi'] > 0 else '#ffffff',
        'color': '#555',
        'weight': 0.5,
        'fillOpacity': 0.9,
    },
    tooltip=folium.GeoJsonTooltip(
        fields=['Nazwa_Wyswietlana', 'Noclegi'],
        aliases=['Powiat:', 'Noclegi:'],
        localize=True
    )
).add_to(m)

colormap.add_to(m)

# Å¹rÃ³dÅ‚o danych
source_html = '''
     <div style="position: fixed; 
     bottom: 10px; right: 10px; width: 150px; height: 30px; 
     z-index:9999; font-size:12px; color: #555; background-color: rgba(255,255,255,0.9); padding: 5px; border-radius: 5px; border: 1px solid #ccc;">
     ðŸ“Š Å¹rÃ³dÅ‚o: BDL GUS
     </div>
     '''
m.get_root().html.add_child(folium.Element(source_html))

plik = f"mapa_turystyczna_integer_{ostatni_rok}.html"
m.save(plik)
print(f"ðŸŽ‰ GOTOWE! Plik: {plik}")
```
